```python
import json
from kafka import KafkaConsumer

# Ustawienia konsumenta
consumer = KafkaConsumer('test', bootstrap_servers='localhost:9092')

# Odbieranie i przetwarzanie wiadomości
for message in consumer:
    if message.value:  # Sprawdzamy, czy wiadomość nie jest pusta
        try:
            data = json.loads(message.value)
            print(f"Otrzymano wiadomość: {data}")
        except json.JSONDecodeError:
            print("Wiadomość nie jest poprawnym JSON-em:", message.value)
    else:
        print("Otrzymano pustą wiadomość")
```
