```python
from binance.client import Client
from kafka import KafkaProducer
import json
import time

# Wprowadź swoje API Key i Secret Key
api_key = 'twoje_api_key'
api_secret = 'twoje_api_secret'

# Inicjalizacja klienta Binance
client = Client(api_key, api_secret)

# Funkcja do pobierania danych o rynku
def get_binance_data():
    # Pobranie ceny bitcoina (BTC/USDT)
    btc_price = client.get_symbol_ticker(symbol="BTCUSDT")
    return btc_price

# Ustawienia połączenia do Kafki
producer = KafkaProducer(bootstrap_servers='localhost:9092', value_serializer=lambda v: json.dumps(v).encode('utf-8'))

# Wysyłanie danych do Kafki
def send_data_to_kafka():
    while True:
        data = get_binance_data()  # Pobieranie danych z Binance
        producer.send('test', data)  # Wysyłanie danych do tematu 'test'
        print(f"Wysłano dane: {data}")
        time.sleep(5)  # Czekaj 5 sekund przed kolejnym pobraniem danych

if __name__ == "__main__":
    send_data_to_kafka()

```
